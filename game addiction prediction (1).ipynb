```python
#which age grp has highest engagement
#this code is gpt generated.
bins = [0, 10, 20, 30, 40, 50, 60]
labels = ['0-10', '11-20', '21-30', '31-40', '41-50', '51-60']
df2=df.copy()
df2['Age Group'] = pd.cut(df2['Age'], bins=bins, labels=labels, right=False)

# Plotting with Seaborn
sns.countplot(x='Age Group', hue='EngagementLevel', data=df2, order=labels, hue_order=['Low', 'Medium', 'High'])
```

```python
#which game genre has highest engagment  GPT GENERATED
genre_engagement = df.groupby('GameGenre')['EngagementLevel'].value_counts(normalize=True).unstack()

# Identify the genre with the highest mean engagement level
highest_engagement_genre = genre_engagement.idxmax(axis=0)['High']
```

Model Building

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.7) #lets go with a 70 30 split
```

Algorithms-> It is a classification problem.
Algorithms to try 
1) Logistic Regression
2) Random Forest
3) Decision Tree
4) KNN

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
Y_predict=lr.predict(X_test)
```

```python
model_accuracy_lr=lr.score(X,Y)
```

```python
model_accuracy_lr
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(Y_test,Y_predict))
```

```python
print(confusion_matrix(Y_test,Y_predict))
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier()
```

```python
rfc.fit(X_train,Y_train)
```

```python
Y_predict_rf=rfc.predict(X_test)
```

```python
rfc.score(X_test,Y_predict_rf)
```

```python
print(classification_report(Y_test,Y_predict_rf))
```

Feature Selection

```python
#using random forest classifier
importances= rfc.feature_importances_
```

```python
final_features=pd.DataFrame({"Features": pd.DataFrame(X).columns, "Importances":importances})
final_features.set_index('Importances')

final_features=final_features.sort_values('Importances')
final_features.plot.bar(color='teal')
```

```python
importances= rfc.feature_importances_
feature_names=X.columns
feature_importances=pd.Series(importances, index=feature_names).sort_values(ascending=False)
```
